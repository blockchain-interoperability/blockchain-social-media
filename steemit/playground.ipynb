```python
from steem import Steem

stm = Steem(nodes=['https://api.steemit.com'])
```

```python
query = {
        "limit":100, #number of posts
        "tag":"" #tag of posts
        }
    #post list for selected query
collected = []
collected_ids = []
cnt = 1
while len(collected) < 10000:
    print(cnt)
    posts = stm.get_discussions_by_created(query)
    new_posts = filter(lambda x: not x['post_id'] in collected_ids, posts)
    collected += new_posts
    collected_ids += map(lambda x: x['post_id'], new_posts)
    print(f'added {len(collected)} posts')
    cnt+=1
```
